```python
import pandas as pd
import seaborn as sns
sns.set_style('whitegrid')

df = pd.read_csv('table.csv', parse_dates=['Date']).set_index('Date')

df.sort_index(inplace=1)

df.head()
```

```python
df['Adj Close'].plot( figsize=(15,5) )
```

```python
t0 = pd.datetime(2008,5,19)
t1 = pd.datetime(2008,9,17) # Margin call

ret01 = ( df.ix[t1]['Adj Close'] - df.ix[t0]['Adj Close'] ) / df.ix[t0]['Adj Close']

print('The return on the investment is {0:1.1f} %'.format( 2 * ret01 * 100 ) )
```

```python
t0 = pd.datetime(2008,5,19)
t2 = pd.datetime(2008,9,19) # Margin call

ret02 = ( df.ix[t2]['Adj Close'] - df.ix[t0]['Adj Close'] ) / df.ix[t0]['Adj Close']

print('The return on the investment is {0:1.1f} %'.format( 2 * ret02 * 100 ) )
```
